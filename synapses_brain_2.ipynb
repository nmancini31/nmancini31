```python
## code to display ALL postsynapses on neuron of interest + presynapses from specific input neurons  (in progress)

import numpy as np
import datetime
import pandas as pd
from caveclient import CAVEclient

datastack_name = "flywire_fafb_production"
client = CAVEclient(datastack_name)

# Define the segments of interest for the pre-synaptic neurons
pre_synaptic_neurons = [720575940631740497]

# Neuron of interest
MDN = [720575940610236514]

# Query output synapses from the MDN neuron
output_synapses_df = client.materialize.query_view("valid_synapses_nt_np", 
                                                   filter_in_dict={"post_pt_root_id": MDN},
                                                   desired_resolution=[4, 4, 40],
                                                   materialization_version=783)

output_synapses_df_filtered = output_synapses_df.groupby('post_pt_root_id').filter(lambda x: len(x) >= 10) # add a filter equal or >5 syn

# Query input synapses from specific upstream neuron to the MDN neuron
input_synapses_df = client.materialize.query_view("valid_synapses_nt_np", 
                                                   filter_in_dict={"pre_pt_root_id": pre_synaptic_neurons, "post_pt_root_id": MDN},
                                                   desired_resolution=[4, 4, 40],
                                                   materialization_version=783)

input_synapses_df = input_synapses_df[input_synapses_df['pre_pt_root_id'].isin(pre_synaptic_neurons)] # ensures that each input synapse is from one of the pre-synaptic neurons specified in pre_synaptic_neurons.
input_synapses_df_filtered = input_synapses_df.groupby('post_pt_root_id').filter(lambda x: len(x) >= 10) # add a filter equal or >5 syn

# Build a NG link that has both pre-syn & post-syn line annotations with pre-syn segments
from IPython.display import HTML
from nglui import statebuilder


# Define image and segmentation layers
img_layer = statebuilder.ImageLayerConfig(client.info.image_source())
seg_layer = statebuilder.SegmentationLayerConfig(client.info.segmentation_source(),
                                                 selected_ids_column='post_pt_root_id')

# Define colors for input and output synapses
input_synapse_color = '#FF0000'  # Red color for input synapses
output_synapse_color = '#0000FF'  # Blue color for output synapses

# Build point mapper for both input and output synapses
input_lines = statebuilder.LineMapper(point_column_a='pre_pt_position', 
                                      point_column_b='pre_pt_position')
output_lines = statebuilder.LineMapper(point_column_a='post_pt_position', 
                                       point_column_b='post_pt_position')

# Create separate annotation layers for input and output synapses
input_anno_layer = statebuilder.AnnotationLayerConfig(name='Input Synapses',
                                                      mapping_rules=input_lines,
                                                      color=input_synapse_color)
output_anno_layer = statebuilder.AnnotationLayerConfig(name='Output Synapses',
                                                       mapping_rules=output_lines,
                                                       color=output_synapse_color)

# Optionally, add output synapses to a different layer
output_seg_layer = statebuilder.SegmentationLayerConfig(client.info.segmentation_source(),
                                                         selected_ids_column='post_pt_root_id')

# Build the StateBuilder with both annotation layers
sb1 = statebuilder.StateBuilder([img_layer, seg_layer, input_anno_layer, output_anno_layer],
                                resolution=[4, 4, 40])

# Optionally, add the output synapses to a different layer
sb2 = statebuilder.StateBuilder([img_layer, output_seg_layer, output_anno_layer],
                                resolution=[4, 4, 40])

# Render the states with the combined synapse data
sb1.render_state(input_synapses_df_filtered, return_as='html')  # Render input synapses
sb2.render_state(output_synapses_df_filtered, return_as='html')  # Render output synapses
```

```python
from IPython.display import display, HTML
from nglui import statebuilder

# Define the segments of interest for the pre-synaptic neurons
pre_synaptic_neurons = [720575940631740497, 720575940617019846, 720575940617208065]  # Add the IDs of neurons you're interested in

# Your segments of interest (post-synaptic neuron)
MDN = [720575940610236514]

# Display HTML outputs for each pre-synaptic neuron's synapses
for i, pre_neuron_id in enumerate(pre_synaptic_neurons):
    # Query synapses from the pre-synaptic neuron to the post-synaptic neuron
    syn_df = client.materialize.query_view("valid_synapses_nt_np", 
                                           filter_in_dict={"pre_pt_root_id": [pre_neuron_id], "post_pt_root_id": MDN},
                                           desired_resolution=[4, 4, 40],
                                           materialization_version=783)

    img_layer = statebuilder.ImageLayerConfig(client.info.image_source())
    seg_layer = statebuilder.SegmentationLayerConfig(client.info.segmentation_source(),
                                                     selected_ids_column='post_pt_root_id')

    lines = statebuilder.LineMapper(point_column_a='post_pt_position', 
                                    point_column_b='post_pt_position')

    # Annotation layer for each pre-synaptic neuron's synapses
    anno_layer = statebuilder.AnnotationLayerConfig(name=f'Synapses_Pre_{i}',
                                                     mapping_rules=lines, 
                                                     color='#0000ff')

    sb = statebuilder.StateBuilder([img_layer, seg_layer, anno_layer], resolution=[4, 4, 40])
    html_output = sb.render_state(syn_df, return_as='html')

    # Display the HTML output
    display(html_output)
```
